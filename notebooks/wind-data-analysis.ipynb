```python
datastream = 'nsainterpolatedsondeC1.c1'
startdate = '2020-08-01'
enddate = '2020-10-15'

# Use ACT to easily download the data.  Watch for the data citation!  Show some support
# for ARM's instrument experts and cite their data if you use it in a publication
result = act.discovery.download_arm_data(username, token, datastream, startdate, enddate)
ds_nsa = act.io.read_arm_netcdf(result)
ds_nsa.clean.cleanup()
ds_nsa
ds = ds_nsa[['potential_temp','bar_pres']]
ds
t = range(0,96480,120) #The 120 is a 2 hour step; can make finer
t_strong_ind = []
t_weak_ind = []
M = []
for i in t:
    if not np.isnan(ds['bar_pres'][i,:].values).all(): #checks if all nan
        h = np.nanargmin(np.abs(ds['bar_pres'][i,:].values-85))
    M.append(ds['potential_temp'][i,h])
```

```python
import xarray as xr
import pandas as pd

time_coords = ds_met1['wdir_vec_mean'].coords['time'].values

M_da = xr.DataArray(M, coords=[('time', time_coords)])


wdir_vec_mean_filtered = ds_met1['wdir_vec_mean'].sel(time=time_coords)
wspd_vec_mean_filtered = ds_met1['wspd_vec_mean'].sel(time=time_coords)


ds_combined = xr.Dataset({
    'wdir_vec_mean': wdir_vec_mean_filtered,
    'wspd_vec_mean': wspd_vec_mean_filtered,
    'M': M_da
})

```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sc = plt.scatter(ds_met1['wdir_vec_mean'], ds_met1['wspd_vec_mean'], c=M, cmap='viridis')

cbar = plt.colorbar(sc)
cbar.set_label('M')

plt.xlabel('wdir_vec_mean')
plt.ylabel('wspd_vec_mean')

plt.title('M Data on wdir_vec_mean and wspd_vec_meanat nsa site')

# Show the plot
plt.show()
```
